```python
# 1st way:
x=np.arange(11)
y1=2*x+3
y2=4*x+5
plt.plot(x,y1,'r--',label='Potato')
plt.plot(x,y2,'go',label='Onion')
# inside lines there are three things : color, stye and marker and they can be in any order.
# eg- here g is color (green), : is style of line, and o is marker points on graph.
plt.legend()
font1={'family':'serif','color':'blue','size':20}
plt.title('Vegetable Prices',fontdict=font1)
plt.xlabel('Quantity')
plt.ylabel('Prices')
plt.grid()
plt.show()
```

```python
# 2nd way:
x=np.arange(11)
y1=2*x+3
y2=4*x+5
# plt.plot(x,y1,'r--',x,y2,'g:^')
plt.plot(x,y1,label='Potato', color='r', linestyle="--", linewidth=2, marker='*', markerfacecolor='b', markersize=10)
plt.plot(x,y2,'g:o',label='Onion')
# inside lines there are three things : color, stye and marker and they can be in any order.
# eg- here g is color (green), : is style of line, and o is marker points on graph.
plt.legend()
font1={'family':'serif','color':'blue','size':20}
plt.title('Vegetable Prices',fontdict=font1)
plt.xlabel('Quantity')
plt.ylabel('Prices')
plt.grid()
plt.show()
```

```python
# 3rd way:
x=np.arange(11)
y1=2*x+3
y2=4*x+5
# plt.plot(x,y1,'r--',x,y2,'g:^')
lines=plt.plot(x,y1,x,y2)
plt.setp(lines[0],label='Potato', color='r', linestyle="--", linewidth=2, marker='*', markerfacecolor='b', markersize=10)
plt.setp(lines[1],'color','g',label='Onion')
plt.legend()
font1={'family':'serif','color':'blue','size':20}
plt.title('Vegetable Prices',fontdict=font1)
plt.xlabel('Quantity')
plt.ylabel('Prices')
plt.grid()
plt.show()
```

```python
# bar graph-
x=np.array([1,2,3])*2
y1=np.array([10,20,30])
y2=np.array([13,22,25])
plt.bar(x,y1,tick_label=['Silver','Gold','Platinum'],width=0.5,label="Previous Year")
plt.bar(x+0.5,y2,width=0.5,label="current year")
plt.style.use('dark_background')
plt.ylabel('Prices')
plt.title('Metal Prices')
plt.legend()
plt.show()
```

```python
# horizontal bar graph -
x=np.array([1,2,3])*2
y1=np.array([10,20,30])
y2=np.array([13,22,25])
plt.barh(x,y1,tick_label=['Silver','Gold','Platinum'],label="Previous Year")
plt.barh(x+0.5,y2,label="current year")
plt.style.use('dark_background')
# plt.ylabel('Prices')
plt.title('Metal Prices')
plt.legend()
plt.show()
```

```python
# histogram -
na1= np.random.randint(30,70,100)
na2=np.random.randint(50,90,100)
plt.hist(na1,label='Class A',alpha=0.5)
plt.hist(na2, label='Class B',alpha=0.5)
# alpha is opacity of graphs
plt.xlim(0,100)
plt.ylim(0,20)
plt.legend()
plt.ylabel('Frequency')
plt.xlabel('Marks')
plt.title('Class Report')
plt.show()
```

```python
x=np.linspace(0,5,1000)
y1=np.sin(2*np.pi*x)
y2=np.cos(2*np.pi*x)

plt.figure(figsize=(10,4))

plt.subplot(211)
# subplot is main plot divided in more than 1 figures and each figure having a subplot
# subplot (rows no. cols no. figure no.)
plt.plot(x,y1)
plt.title("Sin Graph")

plt.subplot(212)
plt.plot(x,y2)
plt.title("Cos Graph")

plt.suptitle("Trignometric Functions")
plt.subplots_adjust(hspace=1)
plt.show()
```

```python
x=np.linspace(0,5,1000)
y1=np.sin(2*np.pi*x)
y2=np.cos(2*np.pi*x)

plt.figure(figsize=(10,4))

ax1=plt.subplot(221)
ax2=plt.subplot(222)
ax3=plt.subplot(223)
ax4=plt.subplot(224)

ax1.plot(x,y1)
ax1.set_title("Sin Graph")

ax2.plot(x,y2)
ax2.set_title("Cos Graph")

plt.suptitle("Trignometric Functions")
plt.subplots_adjust(hspace=1)
plt.show()
```

```python
x=np.linspace(0,5,1000)
y1=np.sin(2*np.pi*x)
y2=np.cos(2*np.pi*x)
y3=np.tan(2*np.pi*x)
y4 = 1/np.sin(2*np.pi*x) 
y5 = 1/np.cos(2*np.pi*x)  
y6 = 1/np.tan(2*np.pi*x) 


plt.figure(figsize=(12,8))
fig,ax=plt.subplots(2,3)
#ax=numpy array (2d list)

ax[0,0].plot(x, y1)
ax[0,0].set_title("Sin Graph")

ax[0,1].plot(x, y2)
ax[0,1].set_title("Cos Graph")

ax[0,2].plot(x, y3)
ax[0,2].set_title("Tan Graph")

ax[1,0].plot(x, y4)
ax[1,0].set_title("Cosec Graph")

ax[1,1].plot(x, y5)
ax[1,1].set_title("Sec Graph")

ax[1,2].plot(x, y6)
ax[1,2].set_title("Cot Graph")

fig.suptitle("Trignometric Functions")
fig.subplots_adjust(hspace=0.5,wspace=0.5)
plt.show()
```
