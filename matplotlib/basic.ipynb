```python
x=np.arange(11)
y1=2*x+3
y2=4*x+5
plt.plot(x,y1,'r--',label='Potato')
plt.plot(x,y2,'go',label='Onion')
plt.legend()
font1={'family':'serif','color':'blue','size':20}
# plt.xlabel('Quantity',fontdict=font1)
# plt.ylabel('Prices',fontdict=font1)
plt.title('Vegetable Prices',fontdict=font1)
plt.xlabel('Quantity')
plt.ylabel('Prices')
# plt.title('Vegetable Prices')
plt.grid()
plt.show()
```

```python
x1=np.arange(11)
y11=x**2
y12=x**2.2
y13=x**2.5
plt.plot(x1,y11,label='sq values')
plt.plot(x1,y12,label='2.2 values')
plt.plot(x1,y13,label='2.5 values')
plt.legend()
font1={'family':'serif','color':'blue','size':20}
plt.title('Mathematical Functions',fontdict=font1)
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.grid()
plt.show()
```
