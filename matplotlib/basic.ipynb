```python
# 1st way:
x=np.arange(11)
y1=2*x+3
y2=4*x+5
plt.plot(x,y1,'r--',label='Potato')
plt.plot(x,y2,'go',label='Onion')
# inside lines there are three things : color, stye and marker and they can be in any order.
# eg- here g is color (green), : is style of line, and o is marker points on graph.
plt.legend()
font1={'family':'serif','color':'blue','size':20}
plt.title('Vegetable Prices',fontdict=font1)
plt.xlabel('Quantity')
plt.ylabel('Prices')
plt.grid()
plt.show()
```

```python
# 2nd way:
x=np.arange(11)
y1=2*x+3
y2=4*x+5
# plt.plot(x,y1,'r--',x,y2,'g:^')
plt.plot(x,y1,label='Potato', color='r', linestyle="--", linewidth=2, marker='*', markerfacecolor='b', markersize=10)
plt.plot(x,y2,'g:o',label='Onion')
# inside lines there are three things : color, stye and marker and they can be in any order.
# eg- here g is color (green), : is style of line, and o is marker points on graph.
plt.legend()
font1={'family':'serif','color':'blue','size':20}
plt.title('Vegetable Prices',fontdict=font1)
plt.xlabel('Quantity')
plt.ylabel('Prices')
plt.grid()
plt.show()
```

```python
# 3rd way:
x=np.arange(11)
y1=2*x+3
y2=4*x+5
# plt.plot(x,y1,'r--',x,y2,'g:^')
lines=plt.plot(x,y1,x,y2)
plt.setp(lines[0],label='Potato', color='r', linestyle="--", linewidth=2, marker='*', markerfacecolor='b', markersize=10)
plt.setp(lines[1],'color','g',label='Onion')
plt.legend()
font1={'family':'serif','color':'blue','size':20}
plt.title('Vegetable Prices',fontdict=font1)
plt.xlabel('Quantity')
plt.ylabel('Prices')
plt.grid()
plt.show()
```

```python
x1=np.arange(11)
y11=x1**2
y12=x1**2.2
y13=x1**2.5
plt.plot(x1,y11,label='sq values')
plt.plot(x1,y12,label='2.2 values')
plt.plot(x1,y13,label='2.5 values')
plt.legend()
font1={'family':'serif','color':'blue','size':20}
plt.title('Mathematical Functions',fontdict=font1)
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.grid()
plt.show()
```

```python
# bar graph-
x=np.array([1,2,3])*2
y1=np.array([10,20,30])
y2=np.array([13,22,25])
plt.bar(x,y1,tick_label=['Silver','Gold','Platinum'],width=0.5,label="Previous Year")
plt.bar(x+0.5,y2,width=0.5,label="current year")
plt.style.use('dark_background')
plt.ylabel('Prices')
plt.title('Metal Prices')
plt.legend()
plt.show()
```

```python
# horizontal bar graph -
x=np.array([1,2,3])*2
y1=np.array([10,20,30])
y2=np.array([13,22,25])
plt.barh(x,y1,tick_label=['Silver','Gold','Platinum'],label="Previous Year")
plt.barh(x+0.5,y2,label="current year")
plt.style.use('dark_background')
# plt.ylabel('Prices')
plt.title('Metal Prices')
plt.legend()
plt.show()
```

```python
# scatter-
x=np.arange(11)
y1=2*x+3
y2=4*x+5
plt.scatter(x,y1,label='Potato', color='r', marker='*')
plt.scatter(x,y2,label='Onion')
plt.legend()
font1={'family':'serif','color':'blue','size':20}
plt.title('Vegetable Prices',fontdict=font1)
plt.xlabel('Quantity')
plt.ylabel('Prices')
plt.grid()
plt.show()
```

```python
# histogram -
na1= np.random.randint(30,70,100)
na2=np.random.randint(50,90,100)
plt.hist(na1,label='Class A',alpha=0.5)
plt.hist(na2, label='Class B',alpha=0.5)
# alpha is opacity of graphs
plt.xlim(0,100)
plt.ylim(0,20)
plt.legend()
plt.ylabel('Frequency')
plt.xlabel('Marks')
plt.title('Class Report')
plt.show()
```
